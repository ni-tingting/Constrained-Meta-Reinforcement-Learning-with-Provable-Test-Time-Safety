```python
import sys
sys.path.append('../..')
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
# Load the data from the pickle file
with open("data/extracted_runs.pkl", "rb") as f:
    data = pickle.load(f)

r_run_all = data["r_run_all"]
c_run_all = data["c_run_all"]
re_avg = np.mean(r_run_all, axis=0)
re_max  = np.quantile(r_run_all, 0.9, axis=0)
re_min = np.quantile(r_run_all, 0.1, axis=0)

c_avg  = np.mean(c_run_all, axis=0)
c_max  = np.quantile(c_run_all, 0.9, axis=0)
c_min  = np.quantile(c_run_all, 0.1, axis=0)
with open("data1/r_c_history_no_safe.pkl", "rb") as f:
    data = pickle.load(f)

r_all_run_no_safe = data["r_all_run_no_safe"]
c_all_run_no_safe = data["c_all_run_no_safe"]

# quantile of the data
re_avg_no_safe = np.mean(r_all_run_no_safe, axis=0)
re_max_no_safe  = np.quantile(r_all_run_no_safe, 0.9, axis=0)
re_min_no_safe = np.quantile(r_all_run_no_safe, 0.1, axis=0)

c_avg_no_safe  = np.mean(c_all_run_no_safe, axis=0)
c_max_no_safe  = np.quantile(c_all_run_no_safe, 0.9, axis=0)
c_min_no_safe  = np.quantile(c_all_run_no_safe, 0.1, axis=0)

with open("data/r_c_history_log_barrier.pkl", "rb") as f:
    data = pickle.load(f)

r_all_run_log_barrier = data["r_log"]
c_all_run_log_barrier = data["c_log"]
# quantile of the data
re_avg_log_barrier = np.mean(r_all_run_log_barrier, axis=0)
re_max_log_barrier  = np.quantile(r_all_run_log_barrier, 0.9, axis=0)
re_min_log_barrier = np.quantile(r_all_run_log_barrier, 0.1, axis=0)

c_avg_log_barrier  = np.mean(c_all_run_log_barrier, axis=0)
c_max_log_barrier  = np.quantile(c_all_run_log_barrier, 0.9, axis=0)
c_min_log_barrier  = np.quantile(c_all_run_log_barrier, 0.1, axis=0)

with open("data/r_c_history_dope_plus.pkl", "rb") as f:
    data = pickle.load(f)

r_all_run_dope_plus = data["r_dope"]
c_all_run_dope_plus = data["c_dope"]
# quantile of the data
re_avg_dope_plus = np.mean(r_all_run_dope_plus, axis=0)
re_max_dope_plus  = np.quantile(r_all_run_dope_plus, 0.9, axis=0)
re_min_dope_plus = np.quantile(r_all_run_dope_plus, 0.1, axis=0)
c_avg_dope_plus  = np.mean(c_all_run_dope_plus, axis=0)
c_max_dope_plus  = np.quantile(c_all_run_dope_plus, 0.9, axis=0)
c_min_dope_plus  = np.quantile(c_all_run_dope_plus, 0.1, axis=0)

with open("data/r_c_history_dope.pkl", "rb") as f:
    data = pickle.load(f)

r_all_run_dope = data["r_dope"]
c_all_run_dope = data["c_dope"]
# quantile of the data
re_avg_dope = np.mean(r_all_run_dope, axis=0)
re_max_dope  = np.quantile(r_all_run_dope, 0.9, axis=0)
re_min_dope = np.quantile(r_all_run_dope, 0.1, axis=0)
c_avg_dope  = np.mean(c_all_run_dope, axis=0)
c_max_dope  = np.quantile(c_all_run_dope, 0.9, axis=0)
c_min_dope  = np.quantile(c_all_run_dope, 0.1, axis=0)
# plot
colorlist = ['tab:blue','tab:green','tab:orange', 'tab:red','tab:purple', 'tab:gray','tab:brown', 'tab:pink']
linestyle_str = ['dotted','dashed', 'dashdot','solid', 'long dash with offset']
# Create (2,2) sub plots
gs = gridspec.GridSpec(2, 1)
fig = plt.figure(figsize=(20,20))

ax = plt.subplot(gs[0, 0])
ax.plot(range(np.size(re_avg[:])), re_avg[:], color = 'tab:green', linestyle = 'solid', label = 'Safe PCE', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(re_avg[:])), re_min[:], re_max[:], alpha = 0.3, facecolor = 'tab:green')

ax.plot(range(np.size(re_avg_no_safe[:])), re_avg_no_safe[:], color = 'tab:blue', linestyle = 'solid', label = 'PCE', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(re_avg_no_safe[:])), re_min_no_safe[:], re_max_no_safe[:], alpha = 0.3, facecolor = 'tab:blue')

ax.plot(range(np.size(re_avg_log_barrier[:])), re_avg_log_barrier[:], color = 'tab:purple', linestyle = 'solid', label = 'LB-SGD', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(re_avg_log_barrier[:])), re_min_log_barrier[:], re_max_log_barrier[:], alpha = 0.3, facecolor = 'tab:purple')

ax.plot(range(np.size(re_avg_dope[:])), re_avg_dope[:], color = 'tab:orange', linestyle = 'solid', label = 'DOPE', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(re_avg_dope[:])), re_min_dope[:], re_max_dope[:], alpha = 0.3, facecolor = 'tab:orange')

ax.plot(range(np.size(re_avg_dope_plus[:])), re_avg_dope_plus[:], color = 'tab:brown', linestyle = 'solid', label = 'DOPE+', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(re_avg_dope_plus[:])), re_min_dope_plus[:], re_max_dope_plus[:], alpha = 0.3, facecolor = 'tab:brown')

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
ax.set_xlabel("Iteration number k", color = 'black', fontsize=22)
ax.set_ylabel("Reward Regret", color = 'black', fontsize=22)
plt.legend(fontsize=20)
plt.legend(loc = 'upper left', fontsize=22)
plt.title('Reward Regret across different environments', fontsize=24)

ax = plt.subplot(gs[1, 0]) 
ax.plot(range(np.size(c_avg[:])), c_avg[:], color = 'tab:green', linestyle = 'solid', label = 'Safe PCE', linewidth=2)
plt.legend(loc = 'center right')
ax.fill_between(range(np.size(c_avg[:])), c_min[:], c_max[:], alpha = 0.3, facecolor = 'tab:green')

ax.plot(range(np.size(c_avg_no_safe[:])), c_avg_no_safe[:], color = 'tab:blue', linestyle = 'solid', label = 'PCE', linewidth=2)
ax.fill_between(range(np.size(c_avg_no_safe[:])), c_min_no_safe[:], c_max_no_safe[:], alpha = 0.3, facecolor = 'tab:blue')
plt.legend(loc = 'lower right')

ax.plot(range(np.size(c_avg_log_barrier[:])), c_avg_log_barrier[:], color = 'tab:purple', linestyle = 'solid', label = 'LB-SGD', linewidth=2)
ax.fill_between(range(np.size(c_avg_log_barrier[:])), c_min_log_barrier[:], c_max_log_barrier[:], alpha = 0.3, facecolor = 'tab:purple')
plt.legend(loc = 'lower right')

ax.plot(range(np.size(c_avg_dope[:])), c_avg_dope[:], color = 'tab:orange', linestyle = 'solid', label = 'DOPE', linewidth=2)
ax.fill_between(range(np.size(c_avg_dope[:])), c_min_dope[:], c_max_dope[:], alpha = 0.3, facecolor = 'tab:orange')
plt.legend(loc = 'lower right')

ax.plot(range(np.size(c_avg_dope_plus[:])), c_avg_dope_plus[:], color = 'tab:brown', linestyle = 'solid', label = 'DOPE+', linewidth=2)
ax.fill_between(range(np.size(c_avg_dope_plus[:])), c_min_dope_plus[:], c_max_dope_plus[:], alpha = 0.3, facecolor = 'tab:brown')
plt.legend(loc = 'lower right')

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
ax.set_xlabel("Iteration number k", color = 'black', fontsize=22)
ax.set_ylabel("Constraint value", color = 'black', fontsize=22)
# ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.axhline(y =  1.5, color = 'tab:red',label = 'Threshold = 1.5', linewidth=2)
ax.legend(fontsize=20)
plt.legend(loc = 'upper right', fontsize=22)
plt.title('Constraint values across different environments', fontsize=24)


# fig.savefig('learning.svg', format='svg', dpi=1200)
plt.tight_layout()
plt.savefig("plots/compare_safe_no_safe_LB_SGD_DOPE_dope_plus.png", format='png', dpi=800)  # Save with 800 dpi
plt.show()
```

```python
from scipy.ndimage import gaussian_filter1d
import matplotlib.ticker as ticker
# Smooth all curves with Gaussian filter
sigma = 10  # adjust for more/less smoothing

re_avg_smooth = gaussian_filter1d(re_avg, sigma)
re_avg_no_safe_smooth = gaussian_filter1d(re_avg_no_safe, sigma)
re_avg_log_barrier_smooth = gaussian_filter1d(re_avg_log_barrier, sigma)
re_avg_dope_smooth = gaussian_filter1d(re_avg_dope, sigma)
re_avg_dope_plus_smooth = gaussian_filter1d(re_avg_dope_plus, sigma)

c_avg_smooth = gaussian_filter1d(c_avg, sigma)
c_avg_no_safe_smooth = gaussian_filter1d(c_avg_no_safe, sigma)
c_avg_log_barrier_smooth = gaussian_filter1d(c_avg_log_barrier, sigma)
c_avg_dope_smooth = gaussian_filter1d(c_avg_dope, sigma)
c_avg_dope_plus_smooth = gaussian_filter1d(c_avg_dope_plus, sigma)

# Plot smoothed curves
gs = gridspec.GridSpec(2, 1)
fig = plt.figure(figsize=(20,20))

ax = plt.subplot(gs[0, 0])
# Use 1-indexed range for log scale (avoid log(0))
x_axis = range(1, len(re_avg_smooth) + 1)
x_axis_fill = range(1, np.size(re_avg[:]) + 1)

ax.plot(x_axis, re_avg_smooth, color='tab:green', linestyle='solid', label='Safe PCE', linewidth=2)
ax.fill_between(x_axis_fill, re_min[:], re_max[:], alpha=0.3, facecolor='tab:green')

ax.plot(x_axis, re_avg_no_safe_smooth, color='tab:blue', linestyle='solid', label='PCE', linewidth=2)
ax.fill_between(x_axis_fill, re_min_no_safe[:], re_max_no_safe[:], alpha=0.3, facecolor='tab:blue')

ax.plot(x_axis, re_avg_log_barrier_smooth, color='tab:purple', linestyle='solid', label='LB-SGD', linewidth=2)
ax.fill_between(x_axis_fill, re_min_log_barrier[:], re_max_log_barrier[:], alpha=0.35, facecolor='tab:purple')

ax.plot(x_axis, re_avg_dope_smooth, color='tab:pink', linestyle='solid', label='DOPE', linewidth=2)
ax.fill_between(x_axis_fill, re_min_dope[:], re_max_dope[:], alpha=0.35, facecolor='tab:pink')

ax.plot(x_axis, re_avg_dope_plus_smooth, color='tab:orange', linestyle='solid', label='DOPE+', linewidth=2)
ax.fill_between(x_axis_fill, re_min_dope_plus[:], re_max_dope_plus[:], alpha=0.35, facecolor='tab:orange')

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
ax.set_xlabel("Iteration number k", color='black', fontsize=22)
ax.set_ylabel("Reward Regret", color='black', fontsize=22)
plt.legend(loc='upper left', fontsize=22)
plt.title('Reward Regret across different environments', fontsize=24)

ax = plt.subplot(gs[1, 0]) 
x_axis_c = range(1, len(c_avg_smooth) + 1)
x_axis_fill_c = range(1, np.size(c_avg[:]) + 1)

ax.plot(x_axis_c, c_avg_smooth, color='tab:green', linestyle='solid', label='Safe PCE', linewidth=2)
ax.fill_between(x_axis_fill_c, c_min[:], c_max[:], alpha=0.3, facecolor='tab:green')

ax.plot(x_axis_c, c_avg_no_safe_smooth, color='tab:blue', linestyle='solid', label='PCE', linewidth=2)
ax.fill_between(x_axis_fill_c, c_min_no_safe[:], c_max_no_safe[:], alpha=0.3, facecolor='tab:blue')

ax.plot(x_axis_c, c_avg_log_barrier_smooth, color='tab:purple', linestyle='solid', label='LB-SGD', linewidth=2)
ax.fill_between(x_axis_fill_c, c_min_log_barrier[:], c_max_log_barrier[:], alpha=0.35, facecolor='tab:purple')

ax.plot(x_axis_c, c_avg_dope_smooth, color='tab:pink', linestyle='solid', label='DOPE', linewidth=2)
ax.fill_between(x_axis_fill_c, c_min_dope[:], c_max_dope[:], alpha=0.35, facecolor='tab:pink')

ax.plot(x_axis_c, c_avg_dope_plus_smooth, color='tab:orange', linestyle='solid', label='DOPE+', linewidth=2)
ax.fill_between(x_axis_fill_c, c_min_dope_plus[:], c_max_dope_plus[:], alpha=0.35, facecolor='tab:orange')

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
ax.set_xlabel("Iteration number k", color='black', fontsize=22)
ax.set_ylabel("Constraint value", color='black', fontsize=22)
ax.axhline(y=1.5, color='tab:red', label='Threshold = 1.5', linewidth=2)

# Format y-axis for Reward Regret (first subplot) to use scientific notation
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_powerlimits((0, 0))
fig.axes[0].yaxis.set_major_formatter(formatter)
plt.legend(loc='upper right', fontsize=22)
plt.title('Constraint values across different environments', fontsize=24)

# Make the scientific notation (e.g. 1e4) bigger for all subplots
for axis in fig.axes:
    # Force x-axis to use 10^3 scale
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_powerlimits((3, 3))
    axis.xaxis.set_major_formatter(formatter)

    axis.yaxis.get_offset_text().set_fontsize(20)
    axis.xaxis.get_offset_text().set_fontsize(20)

plt.tight_layout()
plt.savefig("plots/compare_safe_no_safe_LB_SGD_DOPE_dope_plus_smooth.png", format='png', dpi=800)
plt.show()
```
